During chemical and biochemical processes, maintaining precise control over laboratory environments is critical, as extreme temperatures can quickly destabilize reactions or ruin sensitive biomolecules like enzymes. Because tracking these temperature changes manually across multiple trials is inefficient, an automatic analyzer is useful to instantly identify the extremes of an experiment.

```python
def analyze_enzyme_temp(t1, t2, t3):
    max_temp = max(t1, t2, t3)
    min_temp = min(t1, t2, t3)
    temp_range = max_temp - min_temp
    return max_temp, min_temp, temp_range

reading_1 = 37.2
reading_2 = 38.5
reading_3 = 36.8

high, low, fluctuation = analyze_enzyme_temp(reading_1, reading_2, reading_3)

print("Enzyme Stability Report")
print(f"Maximum Lab Temperature: {high}°C")
print(f"Minimum Lab Temperature: {low}°C")
print(f"Total Temperature Variation: {fluctuation}°C")
```
